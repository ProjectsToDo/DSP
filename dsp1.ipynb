```python
import numpy as np
import matplotlib.pyplot as plt

x_mu = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0, 0, 0]) # do zmiany przy zmianie N
x_mu_7 = x_mu[:7]
N = 11

# =========================
# Macierze indeksów k i n
# =========================
k = np.arange(N).reshape((N, 1))   # kolumna
n = np.arange(N).reshape((1, N))   # wiersz

print("Macierz k:")
print(k)

print("\nMacierz n:")
print(n)

# =========================
# Macierz W (IDFT)
# W[k,n] = exp(j*2πkn/N)
# =========================
W = np.exp(2j * np.pi * k * n / N)

print("\nMacierz W (IDFT):")
print(W)

# =========================
# Synteza IDFT
# x = (1/N) * W * X
# =========================
X_idft = (1/N) * np.dot(W, x_mu)

print("\nWynik IDFT:")
print(X_idft)

# =========================
# Wykres sygnału xµ
# =========================
plt.figure()
plt.stem(np.real(X_idft))
plt.title("Sygnał po IDFT")
plt.xlabel("Próbka n")
plt.ylabel("Amplituda")
plt.grid(True)
plt.show()
```
